## 3. Select specific rows or columns or both 

```python
# returning a specific value
# NB: starts from index is 0
df.loc[100, 'frequency']
```

```python
# get the rows from position 2 to 5
df.iloc[2:5]
```

```python
df.iloc[:,1]
```

```python
# get a specific row
#df.loc[row,column]
df.loc[4, :]
```

```python
# get a subset of rows
df.loc[[0,2,5], :]
```

```python
# returning a specific column 
# these 2 lines are equivalent
df.loc[:,'name']
```

```python
# get a subset of columns
df.loc[:,['name', 'rank']]
```

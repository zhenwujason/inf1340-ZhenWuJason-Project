```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly
import plotly.express as px
```

# **CLEAN TABLE 1**

```python
#Read Table 1 with valid information
Table1 = pd.read_excel(url,sheet_name="Table 1",skiprows=13, header=[1,2])

#Drop duplicate data of country code and country name
Table1 = Table1.drop (columns=['Sort\norder','Major area, region, country or area of destination'])

#Rename columns to simplify the df
Table1.columns = ['Notes','CountryCode','Typeofdata(a)','S1990','S1995','S2000','S2005','S2010','S2015',
                  'M1990','M1995','M2000','M2005','M2010','M2015',
                  'F1990','F1995','F2000','F2005','F2010','F2015']
Table1
```

```python
#problem: Every column is a variable 
#variables includes: country, groups, counts 

clean_Table1 = Table1.melt(id_vars=list(Table1.columns)[:3],
                           value_vars=list(Table1.columns)[3:], 
                           var_name=["Groups"],value_name="Counts")
clean_Table1.head(10)
```

```python
#Groups column contains information of both sex and year

clean_Table1 = (clean_Table1.assign(gender = lambda x: x.Groups.str[0].astype(str), 
                                    year = lambda x: x.Groups.str[1:].astype(str)).drop("Groups", axis=1))
clean_Table1.head(10)
```

```python
#style the table for easier visual

clean_Table1 = (clean_Table1.replace(to_replace = ["S", "M", "F"], value=["Both","Male","Female"]))
clean_Table1 = clean_Table1[["CountryCode","Notes","Typeofdata(a)","year","gender","Counts"]]
clean_Table1 
```

```python
#Read Table 6 with valid information
Table6 = pd.read_excel(url,sheet_name="Table 6", header=[14,15])

#Drop duplicate data of country code and country name
Table6 = Table6.drop (columns=['Sort\norder','Major area, region, country or area of destination'])

#Rename columns to simplify the df
Table6.columns = ['Notes','CountryCode','Type of data (a)','R1990','R1995','R2000','R2005','R2010','R2015',
                  'P1990','P1995','P2000','P2005','P2010','P2015',
                  'C1990-1995','C1995-2000','C2000-2005','C2005-2010','C2010-2015']
Table6
```

```python
#Split Table 6-1: based on its information category by estimate refugee
Table6_Refugee = Table6.loc[:,['Notes','CountryCode','Type of data (a)','R1990','R1995','R2000','R2005','R2010','R2015']]

#problem: Every column is a variable 
#variables includes: country, groups, Estimate Refugee 
Table6_Refugee = Table6_Refugee.melt(id_vars=list(Table6_Refugee.columns)[:3],
                                     value_vars=list(Table6_Refugee.columns)[3:], 
                                     var_name=["Groups"],value_name="Estimate Refugee")

#Groups column contains information of both category and year
Table6_Refugee = (Table6_Refugee.assign(category = lambda x: x.Groups.str[0].astype(str), 
                                        year = lambda x: x.Groups.str[1:].astype(str)).drop("Groups", axis=1))

#style the table for easier visual
Table6_Refugee = Table6_Refugee[["CountryCode","Notes","Type of data (a)","year","Estimate Refugee"]]
Table6_Refugee
```

```python
#Split Table 6-2: based on its information category by estimate refugee rate
Table6_Refugee_rate = Table6.loc[:,['Notes','CountryCode','Type of data (a)','P1990','P1995','P2000','P2005','P2010','P2015']]

#problem: Every column is a variable 
#variables includes: country, groups, Estimate Refugee rate 
Table6_Refugee_rate = Table6_Refugee_rate.melt(id_vars=list(Table6_Refugee_rate.columns)[:3],
                                               value_vars=list(Table6_Refugee_rate.columns)[3:], 
                                               var_name=["Groups"],value_name="Estimate Refugee %")

#Groups column contains information of both category and year
Table6_Refugee_rate = (Table6_Refugee_rate.assign(category = lambda x: x.Groups.str[0].astype(str), 
                                                  year = lambda x: x.Groups.str[1:].astype(str)).drop("Groups", axis=1))

#style the table for easier visual
Table6_Refugee_rate = Table6_Refugee_rate[["CountryCode","Notes","Type of data (a)","year","Estimate Refugee %"]]
Table6_Refugee_rate
```

```python
#Split Table 6-3: based on its information category by the change in the estimated refugee rate
Table6_Refugee_rate_change = Table6.loc[:,['Notes','CountryCode','Type of data (a)','C1990-1995','C1995-2000','C2000-2005','C2005-2010','C2010-2015']]

#problem: Every column is a variable 
#variables includes: country, groups,  Estimate Refugee rate change
Table6_Refugee_rate_change = Table6_Refugee_rate_change.melt(id_vars=list(Table6_Refugee_rate_change.columns)[:3],
                                                             value_vars=list(Table6_Refugee_rate_change.columns)[3:], 
                                                             var_name=["Groups"],value_name="Estimate Refugee % change")

#Groups column contains information of both category and year
Table6_Refugee_rate_change = (Table6_Refugee_rate_change.assign(category = lambda x: x.Groups.str[0].astype(str), 
                                                                year = lambda x: x.Groups.str[1:].astype(str)).drop("Groups", axis=1))

#style the table for easier visual
Table6_Refugee_rate_change = Table6_Refugee_rate_change[["CountryCode","Notes","Type of data (a)","year","Estimate Refugee % change"]]
Table6_Refugee_rate_change
```

# Table 1 Visual

```python
#clean table based on the information needed
new1 = clean_Table1[['CountryCode','year','gender','Counts']]
#select continent data
continent = new1.query('CountryCode == [903,935,908,904,905,909]')
#replace country code to names
continent[['CountryCode']] = continent[['CountryCode']].astype(str)
continent = (continent.replace(to_replace = ["903", "935", "908", "904", "905", "909"], 
                               value=["Africa", "Asia", "Europe", "Latin America and the Caribbean", 
                                      "Northern America", "Oceania"]))
continent
```

```python
continent_fig = px.bar(continent, x="year", y="Counts", color="gender", barmode="group",
                       facet_col="CountryCode",facet_col_wrap=2, height=1000,
                       title="Figure 1: Migrant Stock by Continents from 1990 to 2015")
continent_fig.for_each_annotation(lambda a: a.update(text=a.text.split("=")[-1]))
continent_fig.update_yaxes(showticklabels=True)
continent_fig.show()
```

```python
#INF1340 DATA CELANING
import pandas as pd

undata = pd.ExcelFile("UN_MigrantStockTotal_2015.xlsx")

#there are multiple sheets in one excel, we should read them seperately 
#Every sheets has few rows of data background information,for easy computational purpose,i choose skip those rows 
imsp = pd.read_excel(undata, 'Table 3',skiprows=14) #imsp=International migrant stock as a percentage of the total population
```

```python
#check basic dataset feasture
imsp.head()
```

```python
#check basic dataset feasture
imsp.info()
```

```python
#Principle 1:Column names need to be informative, variable names and not values
#rename unnamed columns to make them as imformative as the rest
imsp=imsp.rename(columns={"International migrant stock as a percentage of the total population (both sexes)":"b1990","Unnamed: 6":"b1995","Unnamed: 7":"b2000","Unnamed: 8":"b2005","Unnamed: 9":"b2010","Unnamed: 10":"b2015",
                        "International migrant stock as a percentage of the total population (male)":"M1990","Unnamed: 12":"M1995","Unnamed: 13":"M2000","Unnamed: 14" : "M2005","Unnamed: 15" : "M2010", "Unnamed: 16" : "M2015",
                        "International migrant stock as a percentage of the total population (female)":"F1990","Unnamed: 18" : "F1995","Unnamed: 19":"F2000","Unnamed: 20":"F2005","Unnamed: 21":"F2010","Unnamed: 22":"F2015"})
imsp
#Principle 3:Variables need to be on cells, not rows and columns
#after rename, columns have some variables in the title row and these are not informative as well
```

```python
#drop the first non-informative rows
imsp= imsp[1:]
imsp
```

```python
#Principle 2:Each column needs to consist of one and only one variable
#columns contains multiple variables need to be melted and rearranged,create temporary columns
imsp1 = imsp.melt(id_vars=["Sort\norder","Major area, region, country or area of destination","Notes","Country code","Type of data (a)"],var_name=["years_renamed"],value_name="International migrant stock as a percentage of the total population(%)")
imsp1
```

```python
#split temporary column to two different columns with one and only variable
imsp1 = imsp1.assign(Gender = lambda x : x.years_renamed.str[0], Year = lambda x : x.years_renamed.str[1:])
imsp1
```

```python
#drop the temporary column years_renamed，rename other columns to be more informative
imsp2=imsp1.drop(['years_renamed'], axis=1,inplace=True)
imsp2=imsp1.replace(to_replace=["b","F","M"],value=["Both","Male","Female"])
imsp2
```

```python
#Pinciple 3:Variables need to be on cells, not rows and columns 
#Variables in this table alreay in cells
```

```python
#Principle 4:each table column needs to have a singular data type 
#check the current data stye 
imsp2.info()
#there are 6 columns are object and 2 columns are float,split them into two tables by data type+unique combination
```

```python
#clean decimal places to integer, back to original number presentation in the excel sheet
imsp2["Sort\norder"]=imsp2["Sort\norder"].astype("int64")
imsp2["Country code"]=imsp2["Country code"].astype("int64")
imsp2.head(10)
```

```python
#double check data type, ready to split into tables by data type following pricinple 4
imsp2.info()
```

```python
imsp3 = imsp2.set_index(["Major area, region, country or area of destination","Notes","Type of data (a)","International migrant stock as a percentage of the total population(%)","Gender","Year"])
imsp3
```

```python
#setting up Maintable for table 3 excel sheet 
Maintable = pd.DataFrame.from_records(
    columns=["ID","Major area, region, country or area of destination","Notes","Type of data (a)","International migrant stock as a percentage of the total population(%)","Gender","Year"],
    data =[
        (a + 1,b,c,d,e,f,g)
        for (a, (b,c,d,e,f,g)) in enumerate(imsp3.index.unique())
    ],
)
```

```python
Maintable.head(30)
```

```python
#setting up Appendix table for table 3 excel sheet 
Appendix = imsp3[["Sort\norder","Country code"]].copy()
Appendix["ID"] = Maintable.set_index(["Major area, region, country or area of destination","Notes","Type of data (a)","International migrant stock as a percentage of the total population(%)","Gender","Year"])
```

```python
Appendix = Appendix.reset_index(drop=True).set_index("ID")
Appendix.head(50)
```

```python
import numpy as np
import pandas as pd
import os
os.chdir('C:/Users/COOPER/Desktop/1340 MID') 
table5 = pd.read_excel("UN_MigrantStockTotal_2015.xlsx",sheet_name = "Table 5", index_col=[0], skiprows=14)
table5.columns = ['Area', 'Notes','Country code', 'Type of data (a)',
       'amountboth1990-1995', 'amountboth1995-2000',
       'amountboth2000-2005', 'amountboth2005-2010', 
       'amountboth2010-2015', 
       'amountmale1990-1995', 'amountmale1995-2000',
       'amountmale2000-2005', 'amountmale2005-2010', 
       'amountmale2010-2015', 
       'amountfemale1990-1995', 'amountfemale1995-2000',
       'amountfemale2000-2005', 'amountfemale2005-2010',
       'amountfemale2010-2015']
table_rename = table5.drop(table5.index[0])
table_rename.drop(columns=['Notes', 'Type of data (a)'], inplace=True)
table_tidy = table_rename.melt(id_vars=["Area","Country code"], var_name=["Year_gender"],value_name="Annual rate")
table_tidy
```
